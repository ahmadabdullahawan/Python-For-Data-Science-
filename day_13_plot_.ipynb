## **Plotting in python**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.survived.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived',data=df)
```

```python
sns.countplot(x='survived',hue='sex',data=df)
```

```python
#scatter plot
plt.scatter(data=df,x='age',y='fare')
```

```python
sns.scatterplot(data=df,x='age',y='fare',hue='pclass')
```

```python
sns.barplot(data=df,x='sex',y='fare',hue='pclass')
```

```python
sns.pairplot(df)
```

```python
df.hist(
    figsize=(10,10),
    bins=10,
    grid=False

)
```

```python
sns.boxplot(data=df,x='sex',y='fare',hue='pclass')
```

```python
sns.boxenplot(data=df,x='sex',y='fare',hue='pclass')
plt.xlabel('sex')
plt.ylabel('fare')
```

```python
sns.stripplot(data=df,x='sex',y='fare',hue='pclass')
```

```python
#heat map
cor=df.select_dtypes(include=np.number).corr() # Select only numeric columns for correlation calculation
plt.figure(figsize=(10,10))
sns.heatmap(cor)
```

```python
sns.violinplot(data=df,x='sex',y='fare',hue='pclass')
```
