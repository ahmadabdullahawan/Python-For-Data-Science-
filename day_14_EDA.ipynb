```python
df2 = df.copy() #Creating a copy of the original dataframe, preventing modifications to the original df
df2['age'] = df2['age'].fillna(df2['age'].mean()) #Applying fillna to the 'age' column of the copied dataframe, df2
```

```python
df2
```

```python
df3=df2.copy()
```

```python
df3.isnull().sum()
```

```python
df3=df3.drop(columns=['deck'])
df3 # Now df3 will contain the modified DataFrame
```

```python
df4=df3.copy()
```

```python
# Get the mode of the 'embark_town' column
mode_value = df4['embark_town'].mode()[0]

# Convert 'embark_town' to category type
df4['embark_town'] = df4['embark_town'].astype('category')

# Add the mode value to the categories if it's not already present
if mode_value not in df4['embark_town'].cat.categories:
    df4['embark_town'] = df4['embark_town'].cat.add_categories([mode_value])

# Now you can safely fill NaN values with the mode
df4['embark_town'] = df4['embark_town'].fillna(mode_value)
```
