**How to rename columns name **?

```python
df=pd.DataFrame({'Yes': [50, 21], 'No': [131, 2],'Name':['A','B']})
df
```

```python
df.rename(columns={'Yes':'A','No':'B','Name':'C'})
```

```python
#2nd method
df.columns=['X','Y','Z']
df
```

REPLACING ANY CHARACTER OR STRING

```python
df.columns=df.columns.str.replace('X','A')
df
```

```python
df.columns=df.columns.str.replace('Y','B')
df
```

```python
df.columns=df.columns.str.replace('Z','C')
df
```
