Splitting a string into multiple columns


```python
import pandas as pd
```

```python
df=pd.DataFrame({'name':['Abdullah Ahmad','Muhammad ALI','Muhammad illyas'],
                 'Location':['Karachi,Pakistan','Lahore,Pakistan','Islamabad,Pakistan']})
df
```

```python
df.name.str.split(' ')
```

```python
df[['First_Name','Last_Name']]=df.name.str.split(' ',expand=True)
df
```

```python
df
```

```python
df[['city name','country name']]=df.Location.str.split(',',expand=True)
df
```

```python
df=df[['First_Name','Last_Name','city name','country name']]
df
```
